```python
import matplotlib.pyplot as plt
import numpy as np
```

Imports needed libraries

```python
x = np.linspace(0, 1, 100)

def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)
```

Created an array x that ranges from 0-1 in 100 steps, and defined functions that return sin(x) and cos(x)

```python
fig, axes = plt.subplots (nrows = 1, ncols = 2, figsize = (10,5))

axes[0].plot(x, sin(x), color='blue')
axes[0].set_title('sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')

axes[1].plot(x, cos(x), color='violet')
axes[1].set_title('cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')

plt.savefig('session_8_plot.pdf')
```

Created a multiplot that has x vs. sin(x) in the left plot, and x vs. cos(x) in the right plot. Labelled both plots with respective functions and saved plot as a pdf.
