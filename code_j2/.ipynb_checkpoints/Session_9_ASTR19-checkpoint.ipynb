```python
import numpy as np
import matplotlib.pyplot as plt
```

Imports needed libraries

```python
ran_nums = np.random.uniform(0, 1, 1000)

plt.hist(ran_nums, bins = 100, color = 'cyan', edgecolor = 'magenta')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 random numbers between 0-1')

plt.savefig('session_9_plot.pdf')
```

Used numpy to pull 1000 random numbers distributed uniformly between [0,1]. Histogramed the random numbers into 100 bins, and plotted the histogram. Labeled my axes and savedthe figure as a PDF.
